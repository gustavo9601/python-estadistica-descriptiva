## Tablas de pivotaje

```python
import numpy as np
import pandas as pd
import seaborn as sb
```

```python
titanic_bd = sb.load_dataset('titanic')
titanic_bd.head(10)
```

```python
# agrupamos por la columna sex, y obtenemos la media para las columnas numericas
titanic_bd.groupby('sex').mean()
```

```python
# agrupamos por la columna sex,solo para la columnas survived
# titanic_bd.groupby('sex')[['survived', 'pclass']].mean()
titanic_bd.groupby('sex')[['survived']].mean()
```

```python
# agrupacion por sex y pclass, obteniendo la media para los sobrevivientes
titanic_bd.groupby(['sex', 'pclass'])[['survived']].mean()
```

```python
# agurpacion por sex y pclass, obteniendo la media para los sobrevivientes, con unstack apila la informacion de forma horizontal
titanic_bd.groupby(['sex', 'pclass'])[['survived']].mean().unstack()
```

```python
# pivot_table('survived', index='sex', columns='pclass')
# pivot_table(columna o [columnas], index= columna o columnas a mostrar como filas, columns= columnas a la derecha)
titanic_bd.pivot_table('survived', index='sex', columns='pclass')
```

```python
# Obteniendo el promedio de genero (sex), mostrando la edad (agre) y los sovrevivientes (survived) por sus clases sociales (pclass)
titanic_bd.pivot_table(['age', 'survived'], index='sex', columns='pclass')


# Analisis
#"""
#La edad promedio de las mujeres de clase 1, es de 34 a√±os , y el porcentaje de sobrevivientes mujes de clase 1 fue del 96%
#"""
```

```python
# Agrupamiento de una columna usando pandas
# Obtendremos la franja de edad a la que pertenece cada persona

# pd.cut(columna de una serie, bins= divisiones de la agrupacion "intervalos", 
# include_lowest=Boolean si requerimos que se incluya el mejor de los intervalos, precision=numeros de presision,
# right=Boolean para definir abre por la izquierda y cierra por la derecha )

age = pd.cut(titanic_bd['age'], bins=[0,18,25,35,50,65,100], include_lowest=True, precision=1, right=False)
age
```

```python
# Al crear la columna age como conjunto de agrupacion
# podemos usarla en una tabla pivot para que cruce los datos

# agrupamos por sex y age "columna de agrupacion"
# mostramos la informacion de los promedios de survived
# mostramos la informacion de sobrevivientes por clase class
titanic_bd.pivot_table('survived', index=['sex', age], columns='class')
```

```python
# Agrupamiento de una columna usando pandas
# Obtendremos en cuartiles, la tarifa agrupada en cuartiles

# pd.cut(columna, q= cantidad de cuantiles, precision= numeros de presicion a generar)

fare = pd.qcut(titanic_bd['fare'], q=4, precision=1 )
fare
```

```python
# misma agrupmacion por cuartiles que la anterior, pero esta vez para cada agrupacion coloca una etiqueta
fare = pd.qcut(titanic_bd['fare'], q=4, precision=1, labels=['Q1', 'Q2', 'Q3', 'Q4'] )
fare
```

```python
titanic_bd.pivot_table('survived', index=['sex', fare], columns='class')
```

```python
titanic_bd.pivot_table('survived', index=['sex', age], columns=[fare, 'pclass'])
```

```python
# Por defecto aplica sobre el primer parametro la funcion mean
# para cambiar el comportamiento del analisis con aggfunc='funcion de agregacion'
titanic_bd.pivot_table('survived', index=['sex', age], columns=[fare, 'pclass'], aggfunc='count')
```

```python
# Aplicando diferentes operaciones a diferentes columnas
# se puede especificar un diccionarion en aggfunc = {'columna': 'operacion', }
titanic_bd.pivot_table( index=['sex'], columns=['class'], aggfunc={'survived': 'sum', 'fare': 'mean'})
```

```python
# Totalizando por filas y columnas 
# margins=True y margins_name='Nombre de las columnas'
titanic_bd.pivot_table('survived', index=['sex'], columns=['class'], margins_name='Totales', margins=True)
```
