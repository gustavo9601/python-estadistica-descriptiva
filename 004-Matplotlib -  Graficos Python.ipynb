```python
def f(x):
    return np.exp(-x)*np.cos(2*np.pi*x)
```

```python
x1 = np.arange(0,5,0.1)
x2 = np.arange(0,5,0.2)

print(x1)

print(x2)
```

```python
plt.figure(1)
plt.subplot(211)  # (211) relacion de graficos en formato filas y columnas
                    # 211  se crearan 2 filas con 1 columna, el ultimo indica el grafico a graficar
plt.plot(x1, f(x1), "ro", x2, f(x2), 'k')

plt.subplot(212)
plt.plot(x2, f(x2), 'g--')


plt.show()
```

```python
plt.figure(1)
plt.subplot(311)  # (311) relacion de graficos en formato filas y columnas
                    # 311  se crearan 3 filas con 1 columna, el ultimo indica el grafico a graficar
plt.plot(x1, f(x1), "ro", x2, f(x2), 'k')

plt.subplot(312)
plt.plot(x2, f(x2), 'g--')

plt.subplot(313)
plt.plot(x2, f(x2*.5), 'p--')

plt.show()
```

# Textos en los graficos

```python
import numpy as np
import matplotlib.pyplot as plt  
mu = 100
sigma = 20

x = mu + sigma * np.random.randn(10000)
x
```

```python
n, bins, patches = plt.hist(x, 50, normed=1, facecolor="g", alpha=0.6) # creamos un grafico de histograma
# .hist(datosX, datosY)

plt.xlabel("Cociente Intelectual")
plt.ylabel("Probabilidad")
plt.title("Histograma de CI")
plt.text(40,0.015, r'$\mu = 100,\ \sigma=20$')  #.text()   permite poner texto en cualquier parte del grafico
plt.axis([0,200,0,0.025])  # .axis([xIncial, xFinal, yInicial, yFinal])   # permite setear el inicio y fin de los ejes
# .text(posicionX. posicionY, texto o crudo para hacer formulas)

plt.grid(True)   # Pone una cuadricula
```

```python
plt.figure(figsize=(10,6), dpi=90)   #crea una figura de (10 en x, 6 en y) para hacerla mas grande
plt.subplot(1,1,1)  # sera solo una figura con una fila, 1 columna, 1er grafico

x = np.arange(0, 10*np.pi, 0.01)
y = np.cos(x)
plt.plot(x, y , lw=2.0)


plt.annotate('Maximo Local', xy = (4*np.pi, 1), xytext= (15,1.5),
            arrowprops = dict(facecolor = "black", shrink = 0.05))

plt.ylim(-2,2)  # expande el marco del grafico para sualizarlo mejor

# .annotate(texto, coordenas en xy donde pondra el texto, ubicacion de la flecha)
# arrowprops (recibe un diccionario con los valores)
# shrink  como se pronunciara la flecha

plt.show()
```

```python
from matplotlib.ticker import NullFormatter   # permite anular las divisiones de uno de los ejes
mu = 0.5
sd = 0.3
y = mu + sd * np.random.randn(500)
y = y[(y>0) & (y<1)]   # filtramos el arreglo validando
y.sort()   # .sort()  ordena el arreglo de menor a mayor, si no se le pasa parametro
x = np.arange(len(y))
 
```

```python
plt.figure(figsize=(10,8))  # hace la figura total de 10 ulgadas de ancho y 8 de alto
plt.subplot(2,2,1)
plt.plot(x,y)
plt.xscale("linear")
plt.xscale("linear")
plt.grid(True)
plt.title("Escala Lineal")

plt.subplot(2,2,2)
plt.plot(x,y)
plt.grid(True)
plt.yscale("log")
plt.title("Escala Logaritmica")

plt.subplot(2,2,3)
plt.plot(x,y - y.mean())   # .mean() Obtiene la media de los elementos del arreglo
plt.grid(True)
plt.yscale("symlog", linthreshy = 0.01)
plt.title("Escala Log Simetrico")

plt.subplot(2,2,4)
plt.plot(x,y)   # .mean() Obtiene la media de los elementos del arreglo
plt.yscale("logit") # logit curvas logisticas
plt.gca().yaxis.set_minor_formatter(NullFormatter())  # formateando el eje Y para que no se superpongan los numeros
plt.grid(True)
plt.title("Escala Logistica")


plt.subplots_adjust(top = 1.5, bottom = 0.08, right = 0.95, hspace=0.25)

plt.show()
```

# Cambios en los ejes

```python
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
S, C = np.sin(x), np.cos(x)   # asignacion de varias variables dinamica para cada variable, 
```

```python
plt.plot(x, S, color="red", linewidth=1.2, linestyle="--", marker="+")
plt.plot(x, C)
# añadiendo lineas al grafico, llamando varias veces el plot, y no dentro de la misma funcion
plt.xlim(-4,4)  # se da en el eje X, un espacio al inicio y al fin
plt.ylim(-1.5,1.5)  # se da en el eje y, un espacio al inicio y al fin
plt.xticks(np.linspace(-4,4,9, endpoint=True))  
# .ticks(arreglo_con_la_numeracion)  controla la distribucion del eje x 
plt.grid(True)

plt.show()
```
