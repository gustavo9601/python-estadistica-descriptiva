# Como crear un data frame

# Manipulacion de data frames

```python
df3 = pd.DataFrame({
    'key' : ['A','B','C','A','B','C'],
    'value1' : range(6),   # range(numeros_a_generar) permite generar numeros aletaorios 
    'value2' : np.random.randint(0,100,6)   # .randint(inicial, final, cantidad)  genera numeros enteros
})
```

```python
df3
```

# Aggregate

```python
df3.groupby('key').aggregate(['min', np.median, np.mean, max])

#.aggregate([funciones_a_ejecutar_en_la_agrupacion])  # permite ejecutar sobre una agrupacion una serie de funciones
# propias de python entre comillas, usando librerias en duro, o palabras reservadas del lenguaje
```

```python
df3.groupby('key').aggregate({
    'value1': 'min',
    'value2' :max
})

# .aggregate({diccionario_con_columnas})  permite definir individualmente por columna que funciones aplicar
```

# Filter

```python
df3.groupby('key').std()   # std() desviacion estandar   
```

```python
def filtrar_funcion(valor):
    return valor['value2'].std()>29.0    # la funcion recibira un array
                                    # seleccionamos el valor que contega ['value2']
                                    # .std()  calculamos la desviacion media
                                    # >29.0  retornara unicamente los valores mayores a 29.0
```

```python
df3.groupby('key').filter(filtrar_funcion)
    # .filter()   permite realizar un filtrado sobre el DataFrame Original
    # recibe un parametro funcion,que tiene la logica de validacion de filtrado
```

# Trasnform

```python
df3.groupby('key').transform( lambda x: x - x.mean() )
# .trasnform( funcion o callback )  # permite editar la vista de los valores de cada llave 
# cada valor, le resta la media y la retorna
```

# Apply

```python
def norm_by_col_2(x):
    x['value1'] /= x['value2'].sum()  
    
    #  /=  => x['value1'] = x['value1'] /=
    # la operacion realizara la asignacion en value1, la divicion por la sumatoria de la agrupacion en value2
    
    return x
```

```python
usando_apply = df3.groupby('key').apply(norm_by_col_2)

# .apply(funcion que retorne un panda)  // permite hacer trasnformacion y agregado a la una vista de la columnas
```

```python
from IPython.display import display
display(df3, df3.groupby('key').sum(), usando_apply)
```

# Division de datos por columnas

```python

import pandas as pd
import numpy as np

df4 = pd.DataFrame({
    'key' : ['A','B','C','A','B','C'],
    'value1' : range(6),   # range(numeros_a_generar) permite generar numeros aletaorios 
    'value2' : np.random.randint(0,100,6)   # .randint(inicial, final, cantidad)  genera numeros enteros
})

df4
```

```python
L1 = [0,1,0,1,2,0]   
L1
```

```python
df4.groupby(L1).sum()  # le pasamos un arreglo, para que realice la agrupacion en base a los elementos del array
```

```python
df4.groupby('key').sum()
```

```python
df5 = df4.set_index('key') # hacemos una creacion de df5 apartir de df4
                            # .set_index('columna') definimos como indice el nombre de la columna del otro dataframe
df5
```

```python
mapping = {
    'A' :'vowel',
    'B' : 'consonant',
    'C' : 'consonant'
}

mapping
```

```python
df5.groupby(mapping).sum()

# para este caso el .groupby() recibe un diccionario que especifica la agrupacion de columnas
# A : vowel , B Y C : consonant
# en ese orden de ideas, realiza otra agrupacion ya no solo por el nombre de la columna sino por la definida en el diccionario
```

```python
# Agrupando por dos valores, funciones, o diccionarios
df5.groupby([str.lower, mapping]).sum()

#.str.lower  permite convertir los indices de letras a minusculas
```

# Graficos desde data frames y exportacion

```python
years = [year for year in range(1996, 2020)]
# le pasamos a la lita un for range que genere numeros de 1996 a 2020 con avance en 1
years
```

```python
deads = [(y + np.random.randint(1,10000,1)[0] ) for y in years]
# creamos una lista a partr de lista years, y con cada valor quie contiene years ejecutamos una funcion (....funcion y retorna un valor)
# round((y + np.random.uniform(0,1) - 1000)
deads
```

```python
df_pruebas = pd.DataFrame({
    'year' : years,
    'deads': deads
})

df_pruebas
```

```python
df_pruebas.plot(x='year', y='deads', figsize=(10,8))

#.plot(x, y, tamaño_grafico)    // permite graficar sin invocar a ploit python
# recibe un arreglo en los ejes, y las demas configuraciones
```

```python
df_pruebas.plot(kind='scatter', x='year', y='deads', figsize=(10,8))  # el parametro kind="tipo_grafico" modifica el tipo de grafico a mostrar
```

```python
grafico1 = df_pruebas.plot(x='year', y='deads', figsize=(10,8)) 
```

```python
figura1 = grafico1.get_figure()  # creamos una variable a partir del plot con el .get_figure()
```
