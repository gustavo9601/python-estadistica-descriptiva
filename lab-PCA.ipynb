```
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler


```

```
iris = sns.load_dataset('iris')
iris.head()
```

```

# Escalando y normalizando a las columnas numericas
scaler = StandardScaler()
scaled = scaler.fit_transform(
    iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].values
    )

```

```
# Matriz de covarianza
covariance_matrix = np.cov(scaled.T)
covariance_matrix
```

```
sns.pairplot(iris)
```

```
# Graficando columnas puntuales
sns.jointplot(x= iris['petal_length'], y=iris['petal_width'])
# Comparando datos normales contra los datos normalizados
sns.jointplot(x = scaled[:, 2], y = scaled[:,3])
```

Descomposicion en vectores y valores propios: https://numpy.org/doc/stable/reference/generated/numpy.linalg.eig.html

```
# np.linalg.eig > Permite calcular la descomposicion de valores propios de la matriz
eigen_values, eigen_vectors = np.linalg.eig(covariance_matrix)
```

```
eigen_values
```

```
eigen_vectors
```

```
variance_explained = []
for i in eigen_values:
    variance_explained.append((i/sum(eigen_values))*100)

# Muestra los porcentajes de los valores de la varianza obtenida
# 72%, 22%...
print(variance_explained)
```

PCA con scikit : https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html

```
from sklearn.decomposition import PCA
# se le define cuanto es el numero de componentes a reducir
pca = PCA(n_components=2)
pca.fit(scaled)
```

```
pca.explained_variance_ratio_
```

```
reduced_scaled = pca.transform(scaled)
reduced_scaled
```

```
# Añadiendo columnas
iris['pca_1'] = scaled[:,0]
iris['pca_2'] = scaled[:,1]
iris
```
